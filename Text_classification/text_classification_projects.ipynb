## Using Beautiful Soup to extract title and lyrics

### Titles for black eye peas

```python
import requests
from bs4 import BeautifulSoup
```

```python
black_eye_peas_title_soup = BeautifulSoup(black_eye_peas_html)
```

```python
print(black_eye_peas_title_soup.prettify())
```

```python
title_cls_container = black_eye_peas_title_soup.find_all(class_ = 'tal qx fsl')
title_cls_container
```

```python

titles_b = []
for td in title_cls_container:
    href_check = td.find('a', href=True)
    #print(href_check)
    if href_check is not None:
        if 'lyric' in href_check['href']:
            titles_b.append(href_check.text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf_vec = TfidfVectorizer()
```

```python
X = black_fifty.lyrics
y = black_fifty['Artist_Name']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 42, test_size = 0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
tf_vec.fit(X_train)
```

```python
X_train_trans = tf_vec.transform(X_train).todense()
```

```python
X_test_trans = tf_vec.transform(X_test).todense()
```

```python
print(tf_vec.vocabulary_)
```

```python
len(tf_vec.vocabulary_)
```

# Train model

## Baseline model: Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
m = LogisticRegression()
```

```python
m.fit(np.array(X_train_trans), y_train)
```

 ## Make predictions



```python
['hey mama so much better i get money']
```

```python
y1_pred = m.predict(np.asarray(X_test_trans))
y1_pred
```

```python
m.predict(tf_vec.transform(['hey mama so much better i get money']))
```

```python
m.predict_proba(tf_vec.transform(['hey mama so much better i get money']))
```

```python
round(m.score(np.asarray(X_train_trans), y_train),3)
```

```python
round(m.score(np.asarray(X_test_trans), y_test),3)
```
